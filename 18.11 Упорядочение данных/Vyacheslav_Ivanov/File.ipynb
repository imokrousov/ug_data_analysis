```python
import numpy as np
import pandas as pd

from io import BytesIO, TextIOWrapper
from zipfile import ZipFile
from urllib.request import urlopen
```

```python
# Создать фрейм данных со столбцами [Название урока, Время начала, Время окончания, ФИО учителя], для сегодняшнего дня

lessons = """08:30–09:15 Алгебра и НА
09:25–10:10 Алгебра и НА
10:30–11:15 Немец. яз.
13:10–13:55 Информатика
14:05–14:50 Информатика
15:05–15:50 Информатика
16:05–16:50 Алгебра и НА""".split("\n")

teachers = """Петрова Ирина Владимировна
Петрова Ирина Владимировна
Харханова Людмила Михайловна
Корнев Андрей Алексеевич
Корнев Андрей Алексеевич
Корнев Андрей Алексеевич
Петрова Ирина Владимировна""".split("\n")

columns = "Название урока, Время начала, Время окончания, ФИО учителя".split(", ")

lessons = [(x[1], *x[0].split("–"), z) for x, z in zip((y.split(" ", 1) for y in lessons), teachers)]

df = pd.DataFrame(lessons, columns=columns)

df
```
